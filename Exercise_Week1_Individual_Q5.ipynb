```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
species={0:'Setosa', 1:'Versicolor', 2:'Virginica'}
vikrantIris=load_iris()
table=pd.DataFrame(vikrantIris.data,columns=vikrantIris.feature_names)
table['Species']=vikrantIris.target
table['Species']=table['Species'].map(species)
```

```python
sepal_length=table['sepal length (cm)']
```

The aforementioned line of code extracts the sepal length from the data.

```python
sample=[]
```

In the aforementioned line of code, we create an empty list.
We will store sample means from the samples here.

```python
for i in range(100):
  sample1=np.random.choice(sepal_length,20,replace=False)
  sample2=sample1.mean()
  sample.append(sample2)
```

for i in range(100): Makes the code run 100 times and each iteration is on random sample.
.
sample1=np.random.choice(sepal_length,20,replace=False) selects samples of 20 each time from sepal_length without repetetion (replace=False)
.
sample2=sample1.mean() computes the mean (sample1.mean()) and stores it in sample2.
.
 sample.append(sample2) the means which were calculated in sample2 are now added to the empty list we created (sample) outside the for loop i.e. after the loop, we will have 100 samples for mean.

```python
plt.hist(sepal_length)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Histogram of Original Sepal Length")
plt.show()
```

In the aforementioned lines and graph:-
What this shows

Shape of the population

Natural variability

Possibly skewed or uneven

```python
plt.hist(sample)
plt.xlabel("Sample Mean of Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Sample Means (n = 20)")
plt.show()
```

What the aforementioned shows:-

Distribution of means

Usually:

More symmetric

Less spread

Bell-shaped

the distribution of sample is more symmetric compared to original sepal length. Which proves the CLT:- The Central Limit Theorem (CLT) states thatif you take sufficiently large random samples from any population (regardless of its original shape, even if skewed), the distribution of the sample means will approximate a normal distribution (bell curve). This fundamental concept allows statisticians to use normal distribution properties for inference, like creating confidence intervals and hypothesis tests, even when the underlying data isn't normal.
