```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
from scipy.stats import pearsonr
from scipy.stats import ttest_ind
species={0:'Setosa', 1:'Versicolor', 2:'Virginica'}
vikrantIris=load_iris()
table=pd.DataFrame(vikrantIris.data,columns=vikrantIris.feature_names)
table['Species']=vikrantIris.target
table['Species']=table['Species'].map(species)
```

```python
r, p = pearsonr(table['sepal length (cm)'], table['petal length (cm)'])
print(r)
```

```python
if r > 0.1:
    relation = "positive"
elif r < -0.1:
    relation = "negative"
else:
    relation = "none"

print("The relationship is:", relation)
```

```python
plt.scatter(table['sepal length (cm)'], table['petal length (cm)'])
```

```python
# Compute slope and intercept
slope, intercept = np.polyfit(table['sepal length (cm)'], table['petal length (cm)'], 1)

# Plot the line
plt.plot(table['sepal length (cm)'], slope*table['sepal length (cm)'] + intercept, color='red')

# Labels
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.title("Scatter Plot with Best-Fit Line")
plt.show()
```

np.polyfit(..., 1)  fits a straight line (degree = 1)

slope*X + intercept = equation of the line

Red line shows linear trend

Scatter points show individual flowers

As x increases, y increases which means as sepal length increases, petal length also increases.

```python
setosa=table[table['Species']=='Setosa']['petal length (cm)']
versicolor=table[table['Species']=='Versicolor']['petal length (cm)']
```

The aforementioned will be inputs for the t-test

```python
t_stat, p_val = ttest_ind(setosa, versicolor)
print("t-statistic:", t_stat)
print("p-value:", p_val)
```

t-test is showing -39.492719391538095 and p-val 5.404910513441677e-62
which means that the mean is significantly different.

```python
setosaW=table[table['Species']=='Setosa']['petal width (cm)']
versicolorW=table[table['Species']=='Versicolor']['petal width (cm)']
virginica=table[table['Species']=='Virginica']['petal width (cm)']
f, p = f_oneway(setosaW, versicolorW, virginica)
print("F-statistic:", f)
print("p-value:", p)
```

Petal width differ significantly across all three species.
