```python
import pandas as pd
from sklearn.datasets import load_iris
species={0:'Setosa', 1:'Versicolor', 2:'Virginica'}
```

```python
vikrantIris=load_iris()
table=pd.DataFrame(vikrantIris.data, columns=vikrantIris.feature_names)
```

```python
table['Species']=vikrantIris.target
table['Species']=table['Species'].map(species)
```

```python
setosa=table[table['Species']=='Setosa']['petal length (cm)']
versicolor=table[table['Species']=='Versicolor']['petal length (cm)']
virginica=table[table['Species']=='Virginica']['petal length (cm)']
```

```python
setosaSkew=setosa.skew()
versicolorSkew=versicolor.skew()
virginicaSkew=virginica.skew()
print("Skewness for all is:-")
print(f"Setosa:- {setosaSkew}")
print(f"Versicolor:- {versicolorSkew}")
print(f"Virginica:- {virginicaSkew}")
```

```python
if setosaSkew > versicolorSkew and setosaSkew > virginicaSkew:
    most_positive = "setosa"
elif versicolorSkew > setosaSkew and versicolorSkew > virginicaSkew:
    most_positive = "versicolor"
else:
    most_positive = "virginica"
if setosaSkew < versicolorSkew and setosaSkew < virginicaSkew:
    most_negative = "setosa"
elif versicolorSkew < setosaSkew and versicolorSkew < virginicaSkew:
    most_negative = "versicolor"
else:
    most_negative = "virginica"

print(f"Most positively skewed species: {most_positive}")
print(f"Most negatively skewed species: {most_negative}")

```
