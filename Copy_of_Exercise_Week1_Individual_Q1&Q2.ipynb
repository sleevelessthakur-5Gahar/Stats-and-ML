To answer the question when mean tells a different story than median is when the data is imbalaced.
Lets say that the data was imbalanced i.e. there was some skew in the data i.e. some values were way too high than the others.
Such values would affect MEAN whereas MEDIAN remains the same.
Because:-
Mean = Sum of all the values in the dataset/number of values in dataset
Mean is affected by the large values in the dataset.
Median:-
If data set has even number of values:-
Eg:- 2,100,100,4
Median= sum of centeral two values in the dataset/2
100+100/2 = 100
If data set has odd number of values:-
1,2,3,4,5
Median= Central value
So, Median = 3.
